```python
import pandas as pd
```

```python
# Read in the wins csv file
csv_df = pd.read_csv("NBA Team Wins.csv")
# Remove NaN data rows
csv_df.dropna(inplace = True)
csv_df
```

```python
# Write to HTML
csv_df.to_html("NBATeamWins.html")
```

```python
# Read in the team names csv file
csvnames_df = pd.read_csv("team_codes_and_names.csv")
# Remove NaN data rows
csvnames_df.dropna(inplace = True)
#Reset Index
csvnames_df.set_index("Team Code", inplace = True)
csvnames_df = csvnames_df.drop("Unnamed: 0",1)
csvnames_df
```

```python
# Write to HTML
csvnames_df.to_html("NBATeamNames.html")
```
